```python
from constrainedSim2 import *
from helpersGeneration import *


%load_ext autoreload
%reload_ext autoreload
%autoreload 2
# Tests on standard settings:
T = 5000
D = 1
alpha = 1
deltaT = 1
# number of subposition per Frame for image Generation
nPosFrame = 50
l_IBM,h_IBM, w_IBM  = 550,300, 7
l_cri, w_cri = 200,30

rects = []

# Create a rectangle
rects.append(Rectangle(length=l_IBM, width=w_IBM, angle=np.pi/2, center_x=0, center_y=0))
rects.append(Rectangle(length=l_IBM, width=w_IBM, angle=np.pi/2, center_x=0, center_y=h_IBM - w_IBM))

rects.append(Rectangle(length=h_IBM, width=7, angle=0, center_x=(-l_IBM+w_IBM)/2, center_y=(h_IBM-w_IBM)/2))
rects.append(Rectangle(length=h_IBM, width=7, angle=0, center_x=(l_IBM-w_IBM)/2, center_y=(h_IBM-w_IBM)/2))

cri_space = 90
cri_start_x = -l_IBM/2 + cri_space

rects.append(Rectangle(length=l_cri, width=w_cri, angle=0, center_x=cri_start_x + cri_space*0, center_y=(l_cri - w_IBM)/2))
rects.append(Rectangle(length=l_cri, width=w_cri, angle=0, center_x=cri_start_x + cri_space*2, center_y=(l_cri - w_IBM)/2))
rects.append(Rectangle(length=l_cri, width=w_cri, angle=0, center_x=cri_start_x + cri_space*4, center_y=(l_cri - w_IBM)/2))


rects.append(Rectangle(length=l_cri, width=w_cri, angle=0, center_x=cri_start_x + cri_space * 1, center_y=h_IBM - (l_cri + w_IBM)/2))
rects.append(Rectangle(length=l_cri, width=w_cri, angle=0, center_x=cri_start_x + cri_space * 3, center_y=h_IBM - (l_cri + w_IBM)/2))

a = 5
max_dims = [-l_IBM/2 -a, l_IBM/2+a, 0-a, h_IBM + a]

traj = traj_in_rectangles(T,D,alpha,deltaT, rects, startPos=(0,0))

animate_trajectory_with_shapes2(traj,rects,T//nPosFrame,nPosFrame,max_dims= max_dims, figsize = (15,10))
```
